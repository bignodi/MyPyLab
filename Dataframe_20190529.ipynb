```python
import pandas as pd

dicts = {
    "factory":"Taipei",
    "sensor1":"1",
    "sensor2":"2",
    "sensor3":"3",
    "sensor4":"4",
    "sensor5":"5",
}

select = pd.Series(dicts, index = dicts.keys())
print(select)
print("#"*50)

## 索引值取值
print(select[0])
print("#"*50)

## 索引值取值
print(select[[0,2,5]])
print("#"*50)

## Key值取值
print(select[['factory','sensor1','sensor3']])
print("#"*50)


## 資料切片, 取到index 2之前（ 0-1）
print(select[:2])
print("#"*50)

## 資料切片, 取Key值=sensor2 之後所有的值
print(select["sensor2":])
```

```python
## DataFrame 操作

groups = ['Movies','Sports','Coding','Fishing','Dancing','Cooking']
num = [46, 8, 12, 12, 6, 58]

dicts = {
    
    "groups": groups,
    "num": num
}

select_df = pd.DataFrame(dicts)
print(select_df)
print("="*50)

## 回傳列數與欄數
print(select_df.shape)
print("="*50)

## 回傳描述性統計
print(select_df.describe())
print("="*50)

## 回傳前 n 筆
print(select_df.head(3))
print("="*50)

## 回傳最後 n筆
print(select_df.tail(3))
print("="*50)

## 回傳欄位名稱
print(select_df.columns)
print("="*50)

## 回傳 index
print(select_df.index)
print("="*50)

## 回傳資料內容
print(select_df.info)
print("="*50)
```

```python
## DataFrame 資料選擇與篩選

groups = ['Movies','Sports','Coding','Fishing','Dancing','Cooking']
num = [46, 8, 12, 12, 6, 58]

dicts = {
    "groups":groups,
    "num":num
}


select_df2 = pd.DataFrame(dicts)
print(select_df2)
print("="*50)

## 取第一列第二欄 的值
print(select_df2.iloc[0,1])
print("="*50)

## 取第一列 的組名與人數
print(select_df2.iloc[0:1,:])
## 取第二列 的組名與人數
print(select_df2.iloc[1:2,:])
## 取第四列 的組名與人數
print(select_df2.iloc[3:4,:])
#print(select_df2.iloc[3:5,:])  #取了3,4筆
print("="*50)

## 取各組人數， 第二欄
print(select_df2.iloc[:,1])
## 取各組人數， 第一欄
#print(select_df2.iloc[:,0])
print("="*50)


## 取各組人數，指定Key值
print(select_df2["num"])
print("="*50)

## 取各組人數，指定num 欄位
print(select_df2.num)
print("="*50)

## 布林值篩選
#out_df = select_df2
out_df = select_df2[select_df2.loc[:,"num"] > 10]
print(out_df)
```

```python
## df 處理遺漏值

import pandas as pd

##讀檔
df = pd.read_csv('shop_list2.csv')

## 放入 dataframe
select_df = pd.DataFrame(df)

# ix is deprecated 過濾空值
#print(select_df.ix[:,"shop name"].isnull())

## dropna() 有空值都刪除
drop_value = select_df.dropna()
print(drop_value)
print("="*50)

## fillna() 有空值都補0
fill_value = select_df.fillna(0)
print(fill_value)
print("="*50)

## fillna() 依欄位填預設值
fill_default = select_df.fillna({"shop name":"NULL", "market size":"0"})
print(fill_default)
```
